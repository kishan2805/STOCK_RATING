# Income Statement


```python
import requests
import pandas as pd

# Fetch the income statement data
url = 'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol=ibm&apikey=LKY3X6NTHPBSVC67'
r = requests.get(url)
data = r.json()

# Extract the income statement data
income_statement = data.get('annualReports', [])

# Normalize the JSON data to create a DataFrame
df = pd.json_normalize(income_statement)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# List of columns to keep
columns_to_keep = [
    'fiscalDateEnding', 'grossProfit', 'totalRevenue', 'costOfRevenue', 
    'operatingIncome', 'netIncome', 'operatingExpenses', 'incomeBeforeTax', 
    'incomeTaxExpense', 'ebit', 'ebitda', 'researchAndDevelopment'
]

# Filter the DataFrame to keep only the desired columns
df_filtered = df[columns_to_keep]

# Display the filtered DataFrame
print("\nFiltered DataFrame:")
print(df_filtered)

# Convert specified columns to numeric and handle errors (e.g., 'None' values)
numeric_columns = [
    'grossProfit', 'totalRevenue', 'costOfRevenue', 
    'operatingIncome', 'netIncome', 'operatingExpenses', 'incomeBeforeTax', 
    'incomeTaxExpense', 'ebit', 'ebitda', 'researchAndDevelopment'
]

# Convert to numeric and handle errors (e.g., 'None' values)
df_filtered[numeric_columns] = df_filtered[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Ensure 'fiscalDateEnding' remains an object
df_filtered['fiscalDateEnding'] = df_filtered['fiscalDateEnding'].astype(str)

# Display the data types after conversion
print("\nData types after conversion:")
print(df_filtered.dtypes)

# Display the final DataFrame
print("\nFinal DataFrame:")
print(df_filtered)
```
